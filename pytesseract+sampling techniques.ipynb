```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations
```

```python
df = pd.DataFrame({'Name':['Mr.Snow','Ms.Tolson','Mr.Kraft','Ms.Irwin','Mr.Jones'],'Years':[20,22,26,24,28]})
df
```

```python
# Assuming df is your DataFrame with 5 rows
# You can replace it with your actual DataFrame

# Generate all possible combinations of 2 elements
sample_combinations = list(combinations(df.Years, 2))
# Calculate the mean for each combination
sample_means = [sum(comb)/len(comb) for comb in sample_combinations]
# Convert the list of means to a pandas Series
sample_means_series = pd.Series(sample_means)
print(f'MUxbar = {sample_means_series.mean()}')
# Display the distribution of sample means
value_counts = sample_means_series.value_counts(sort=False,normalize=False)

# Plot a bar chart
plt.figure(figsize=(10, 6))
value_counts.plot(kind='bar')
plt.title('Value Counts')
plt.xlabel('Unique Values')
plt.ylabel('Counts')
plt.show()
```

```python
# Define the population
population = [12, 12, 14, 16]

# Define the size of combinations
r = 2

# Generate combinations
comb = list(combinations(population, r))

# Print information about combinations
print(f'The number of all possible samples from the population is: {len(comb)}')
print('All the possible combinations are:', comb)

# Calculate sample means
sample_means_series = pd.Series([sum(i) / len(i) for i in comb])

# Calculate sample distribution of the mean
sample_distribution_of_the_mean = sample_means_series.value_counts(sort=False, normalize=True)

# Plot the sample distribution
sample_distribution_of_the_mean.plot(kind='bar')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.title('Sample Distribution of the Mean')
plt.show()
```

```python
population_mean = sum(population)/len(population)
muxbar = sample_means_series.mean()
print(population_mean)
print(muxbar)
```
